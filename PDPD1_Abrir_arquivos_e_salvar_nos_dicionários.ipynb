```python
dir_list= dict()

# Python program to explain os.listdir() method  
    
# importing os module  
import os 
  
# Get the list of all files and directories 
# in the root directory 

path = "/content/drive/MyDrive/PDPD/PDPD-Arquivos"
dir_list[0] = os.listdir(path) 
  
print("Files and directories in '", path, "/content/drive/MyDrive/PDPD/PDPD-Arquivos")  
  
# print the list 
print(dir_list)

hickle.dump(dir_list, '/content/drive/MyDrive/PDPD/Dicionário/dir_list.hkl', mode='w')
```

```python
#abrindo os aquivos e add no dicionario "dados"
dados = dict()
suj = '25'
k = 0
for j in range(len(dir_list[0])) :
  if dir_list[0][j][16]+dir_list[0][j][17] == suj :
    x = dir_list[0][j]
    y = '/content/drive/MyDrive/PDPD/PDPD-Arquivos/' + x
    dados[k] = loadmat(y)
    k = k +1
create_ds_args = {'compression': "gzip",
                  'shuffle': True,
                  'fletcher32': True}

dictdump.dicttoh5(dados, "/content/drive/MyDrive/PDPD/Dicionário/dados" + suj + ".h5", h5path="/",
          create_dataset_args=create_ds_args)
del dados
```
